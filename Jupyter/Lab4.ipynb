#### **Imprimir cantidad de filas duplicadas y eliminarlas**

```python
duplicados = datos.duplicated().sum()
if duplicados > 0:
  print(f"Se encontraron {duplicados} filas duplicadas, Procediendo a eliminar")
  datos.drop_duplicates(inplace=True)
else:
  print("No se encontraron filas duplicadas")

print("Tras el paso de duplicados:", datos.shape)
```

#### **Eliminar valores nulos**

```python
data_sin_nulos = datos.dropna()
data_sin_nulos.info()
```

#### **Filtrar pokemons tipo fuego**

```python
def filtro_fuego(Type):
  return 'Fire' in Type

datos_fuego = datos[datos['Type 1'].apply(filtro_fuego)]
datos_fuego.head(20)
```

#### **Promedio de defensa de los pokemons tipo fuego**

```python
promedio_defensa = np.mean(datos_fuego['Def'])
promedio_defensa
```

#### **Grafico de barra top 10 nivel ataque tipo fuego**

```python
data_top = datos_fuego.head(10)
data_top.set_index('Name', inplace=True)
data_top['Att'].plot(x=None, kind='bar')
```
